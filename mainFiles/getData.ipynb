```python
import requests as rq
import json
import pandas as pd
import numpy as np
```

```python
link = "https://www.skyblock.bz/api/product/historical/ENCHANTED_QUARTZ"
keyId = "5f4b7a9e-085d-4507-ae07-1cb4b9186f16"
header = {"referer": "https://www.skyblock.bz/"}
x = rq.get(link, headers = header)
```

```python
# turn x into a json object
data = json.loads(x.content)
```

```python
buy_price = []
sell_price = []
day = []
for i in data:
    buy_price.append(i['buy_price'])
    sell_price.append(i['sell_price'])
    day.append(i['day'])

import matplotlib.pyplot as plt
```

```python
# find the top 10 highest buy prices
top_10_buy = np.array(buy_price).argsort()[-10:][::-1]
top_10_buy
```

```python
# plot buy_price and sell_price as y and day as x
plt.plot(day, buy_price, label = 'buy_price')
```
